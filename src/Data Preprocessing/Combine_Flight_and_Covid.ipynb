```python
import numpy as np
import pandas as pd
from datetime import timedelta
```

```python
flight = pd.read_csv('OSN_Jun2021_Clean.csv')
covid = pd.read_csv('JH_time_series_covid19_confirmed_global.csv').fillna(0)
```

## Prepare Flight Data

```python
flight_clean = flight[['callsign', 'airline', 'country', 'origin', 'destination', 'lastseen']]

flight_clean = flight_clean[flight_clean['airline'] != '0']

flight_clean
```

```python
flight_by_country = flight_clean.sort_values(['lastseen', 'country']).groupby([pd.to_datetime(flight_clean['lastseen']).dt.date, 'country']).count()[1:]
```

```python
flight_by_country = flight_by_country[['callsign']]

flight_by_country.reset_index(level = 1, inplace=True)

flight_by_country.reset_index(level = 0, inplace=True)

flight_by_country
```

## Prepare COVID Data

```python
covid_apr = covid.loc[:, '6/1/21':'6/30/21']
```

```python
covid_apr.insert(0, 'Country', covid['Country/Region'])
```

```python
covid_apr.reset_index(drop=True, inplace=True)
```

```python
covid_apr
```

```python
for i in range(covid_apr.shape[0]):
    if covid_apr['Country'][i] == 'Taiwan*':
        covid_apr['Country'][i] = 'Chinese Taipei'
    elif covid_apr['Country'][i] == 'Tanzania, United Republic of':
        covid_apr['Country'][i] = 'Tanzania'
    elif covid_apr['Country'][i] == 'US':
        covid_apr['Country'][i] = 'United States'
    elif covid_apr['Country'][i] == 'Russia':
        covid_apr['Country'][i] = 'Russian Federation'
    elif covid_apr['Country'][i] == 'Korea, South':
        covid_apr['Country'][i] = 'Korea'
    elif covid_apr['Country'][i] == 'China':
        covid_apr['Country'][i] = 'China (People\'s Republic of)'
    elif covid_apr['Country'][i] == 'Czechia':
        covid_apr['Country'][i] = 'Czech Republic'

# covid_apr = covid_apr.rename(columns = {: 'Chinese Taipei', 'Tanzania, United Republic of': , 'US': ,
#                             : ,  : ,  : })
```

```python
cov_clean = covid_apr.groupby(['Country']).sum()[1:]
cov_clean.reset_index(level = 0, inplace=True)
cov_clean
```

## Combine Flight and COVID Data

```python
country_list = []
cov_cnt_list = []
dt_list = []

for i in range(flight_by_country.shape[0]):
    for j in range(cov_clean.shape[0]):
        cov_count = 0
        if flight_by_country['country'][i] == cov_clean['Country'][j]:
            #print(flight_by_country['country'][i])
            
            fl_curr_day = pd.to_datetime(flight_by_country['lastseen'][i]).strftime('%#m/%#d/%y')
            dt_list.append(flight_by_country['lastseen'][i])
            if fl_curr_day == '6/1/21':
                fl_nxt_day = pd.to_datetime(flight_by_country['lastseen'][i] + timedelta(days=1)).strftime('%#m/%#d/%y') 
                #print(fl_nxt_day)
                cov_count = int(cov_clean[cov_clean['Country'] == flight_by_country['country'][i]][fl_nxt_day])
                #print(cov_count)
                cov_cnt_list.append(cov_count)
                country_list.append(flight_by_country['country'][i])
                break
            else:
                fl_prev_day = pd.to_datetime(flight_by_country['lastseen'][i] - timedelta(days=1)).strftime('%#m/%#d/%y') 
                cov_count = int(cov_clean[cov_clean['Country'] == flight_by_country['country'][i]][fl_curr_day]) - int(cov_clean[cov_clean['Country'] == flight_by_country['country'][i]][fl_prev_day])
                cov_cnt_list.append(cov_count)
                country_list.append(flight_by_country['country'][i])
                break
```

```python
cov_apr_list = list(zip(cov_cnt_list, dt_list, country_list))
```

```python
df = pd.DataFrame(cov_apr_list)
```

```python
combined =pd.merge(flight_by_country, df,  how = 'left', left_on = ['lastseen','country'], right_on = [1,2])

final = combined[['lastseen', 'country', 'callsign', 0]]
combined

final = final.rename(columns = {'lastseen': 'Timestamp', "country": "Country", 'callsign' : 'Num_Flights', 0 : 'Num_Covid_Cases'})

final.to_csv('Combined_Data_Jun21.csv', index = False)  
```
