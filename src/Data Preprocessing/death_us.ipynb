## What's the number of death in US since Covid-19's outbreak?

### Goal:
- Final processed dataframe meets the protocol for using influx-client API for importing to InfluxDB.
- Basic time series visualization can be conducted based on this processed result

```python
death = pd.read_csv('time_series_covid19_deaths_global.csv')
death
```

```python
death_us = death[death['Country/Region'] == 'US']
death_us = death_us.T
death_us = death_us.iloc[4:,0]
death_us
```

```python
death_us.plot(figsize=(16, 4), label='Number of Deaths', title='Number of Deaths over time in US', fontsize=14, alpha=0.6)
# plot moving average
#avg_ts.plot(label='Average number of reviews', fontsize=14)
plt.legend(fontsize=14);
```
