## What's the number of death in US since Covid-19's outbreak?

### Goal:
- Final processed dataframe meets the protocol for using influx-client API for importing to InfluxDB.
- Basic time series visualization can be conducted based on this processed result

```python
recover = pd.read_csv('time_series_covid19_recovered_global.csv')
```

```python
recover_us = recover[recover['Country/Region'] == 'US']
recover_us = recover_us.T
recover_us = recover_us.iloc[4:,0]
recover_us
```

```python
recover_us.plot(figsize=(16, 4), label='Number of Recovered', title='Number of Recovered over time in US', fontsize=14, alpha=0.6)
# plot moving average
#avg_ts.plot(label='Average number of reviews', fontsize=14)
plt.legend(fontsize=14);
```
